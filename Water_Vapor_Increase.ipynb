```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
e0 = 0.611  # kPa, reference saturation vapor pressure at T0
T0 = 273.15  # K, reference temperature
Lv = 2.5e6  # J/kg, latent heat of vaporization
Rv = 461  # J/(kg*K), specific gas constant for water vapor

# Mean annual temperatures (converted to Kelvin)
T_honolulu = 23.5 + 273.15  # K
T_utqiagvik = -10.2 + 273.15  # K

# Temperature increases (0 to 5C from current values)
delta_T = np.arange(0, 6, 1)  # 0 to 5C

# Function to compute saturation vapor pressure (kPa)
def saturation_vapor_pressure(T):
    return e0 * np.exp((Lv / Rv) * (1/T0 - 1/T))

# Compute vapor pressures for both locations
es_honolulu = saturation_vapor_pressure(T_honolulu + delta_T)
es_utqiagvik = saturation_vapor_pressure(T_utqiagvik + delta_T)

# Compute absolute and percentage changes
delta_es_honolulu = es_honolulu - es_honolulu[0]
delta_es_utqiagvik = es_utqiagvik - es_utqiagvik[0]

percent_change_honolulu = (delta_es_honolulu / es_honolulu[0]) * 100
percent_change_utqiagvik = (delta_es_utqiagvik / es_utqiagvik[0]) * 100

# Plot results
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot absolute increase in vapor pressure (kPa)
ax[0].plot(delta_T, delta_es_honolulu, label="Honolulu, HI", marker="o")
ax[0].plot(delta_T, delta_es_utqiagvik, label="Utqiagvik, AK", marker="s")
ax[0].set_xlabel("Temperature Increase (C)")
ax[0].set_ylabel("Increase in Vapor Pressure (kPa)")
ax[0].set_title("Additional Water Vapor in kPa")
ax[0].legend()
ax[0].grid(True)

# Plot percentage increase in vapor pressure
ax[1].plot(delta_T, percent_change_honolulu, label="Honolulu, HI", marker="o")
ax[1].plot(delta_T, percent_change_utqiagvik, label="Utqiagvik, AK", marker="s")
ax[1].set_xlabel("Temperature Increase (C)")
ax[1].set_ylabel("Percentage Increase (%)")
ax[1].set_title("Percentage Increase in Water Vapor")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

The increase in Hawaii's water vapor pressure is significantly higher than Alaska's. Nonetheless, there is still a slight increase in vapor pressure for Alaska.

You can see in the percentange increase plot that Alaska has the relative higher increase. This is due to lower temperatures hold less water vapor, so the increase is a large proportion of the total space.
